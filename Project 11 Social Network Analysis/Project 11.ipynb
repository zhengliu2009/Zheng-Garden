```python
import numpy as np
A = np.mat("0 1 0; 0 0 1; 2 1 -2")
print("A=", A)

print("Eigenvalues: ",np.linalg.eigvals(A))

eigenvalue, eigenvector = np.linalg.eig(A)
for i in range(0, len(eigenvalue)):
  print('Eigenvalue {:.2f} has eigenvector {}.'.format(eigenvalue[i], eigenvector[i]))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G1=nx.DiGraph()
G1.add_nodes_from([1,2,3,4,5])
G1.add_edges_from([(1,2),(2,4),(3,1),(3,2),(4,1),(4,5),(5,1)])
nx.draw(G1,with_labels=True)
plt.draw
plt.show()
H1,A1 = nx.hits(G1)
print(A1)
print(H1)
```

Hyperlink-Induced Topic Search (HITS; also known as hubs and authorities) is a link analysis algorithm that rates Web pages, developed by Jon Kleinberg. The idea behind Hubs and Authorities stemmed from a particular insight into the creation of web pages when the Internet was originally forming; that is, certain web pages, known as hubs, served as large directories that were not actually authoritative in the information that they held, but were used as compilations of a broad catalog of information that led users direct to other authoritative pages. In other words, a good hub represented a page that pointed to many other pages, and a good authority represented a page that was linked by many different hubs.

The scheme therefore assigns two scores for each page: its authority, which estimates the value of the content of the page, and its hub value, which estimates the value of its links to other pages.

In other words, Authority is a page with many in-links; A hub is a page with many out-links

In the problem above, Point 1 has the highest Authority score and the lowest Hub score, while Point 3 has the highest Hub score and the lowest Authority score. This is consistent with their notions.

```python
G2=nx.DiGraph()
G2.add_nodes_from([1,2,3,4])
G2.add_edges_from([(1,2),(1,3),(2,3),(2,4),(3,1),(3,4),(4,3)])
nx.draw(G2,with_labels=True)
plt.draw
plt.show()
H2,A2 = nx.hits(G2)
print(A2)
print(H2)
```

In the problem above, we notice that Point 3 has the highest Authority score and the lowest Hub score; while Point 2 has the highest Hub score, Point 1 has the lowest Authority score. After futher examing the figure, we noticed that Point 1 and Point both have 1 IN and 2 OUTs. Yet Point 2 has a lower Hub score than Point 1. This is not consistent with the notion of Authority and Hub. The reason might be that Point 1 has a bi-diretion link with Point 3, which is the Authority of this figure; while Point 2 only has an out link to Point 3.
